## Missing Value analysis 
- As we can see, out of 86 entries in the dataset, there are 80 missing values in the **Average Price Groceries** column.
-  So we have decided to drop this column .

```python
df=df.drop(['Average Price Groceries'],axis=1)
df.head()
```

```python
df.isnull().sum(axis=1)
```

```python
import pandas as pd

numeric_cols = ['Population Density', 'Working Age Population ', 'Unemployment Rate', 'GDP per Capita']

df[numeric_cols] = df[numeric_cols].interpolate(method='linear')

df['Main Spoken Languages'] = df['Main Spoken Languages'].fillna("Unknown")

print(df.isnull().sum())

df.head()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.head(5)
```

```python
df = df.copy()  
df['City'] = df['City'].astype(str).str.strip()  
split_df = df['City'].str.replace(";", ",", regex=False).str.split(",", expand=True)
print(split_df.head())
```

```python
split_df.columns = ['CityName', 'Country']
df = pd.concat([df, split_df], axis=1)
```
